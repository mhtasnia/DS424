```python
import cv2
import numpy as np
image_1 = cv2.imread('/content/drive/MyDrive/download.jpg', cv2.IMREAD_GRAYSCALE)
unique_gray_levels = np.unique(image_1)
print(f"The number of unique gray levels in the image is: {len(unique_gray_levels)}")
```

```python
len(unique_gray_levels)
```

## plot histogram of image

```python
plt.subplot(1,2,1)
plt.imshow(image_1,cmap='gray')
plt.title('image')
plt.subplot(1,2,2)
hist,bins=np.histogram(image_1.ravel(),253,[0,256])
plt.plot(hist)
plt.title('histogram')
plt.grid(True)
plt.show()
```

```python
hist, bins = np.histogram(image_1.ravel(), 256, [0, 256])
cumulative_hist = np.cumsum(hist)
plt.plot(cumulative_hist)
plt.xlabel('Gray Level')
plt.ylabel('Cumulative Frequency')
plt.title('Cumulative Histogram')
plt.grid(True)
plt.show()
```

# **Shifting**

```python
image_1_shifted = np.clip(image_1 - 100, 0, 255).astype(np.uint8)
```

```python
plt.imshow(image_1_shifted,cmap='gray')
plt.axis(False)
```

```python
unique_gray_levels_shifted = np.unique(image_1)

print(f"The number of unique gray levels in the image is: {len(unique_gray_levels_shifted)}")
```

```python
plt.subplot(1,2,1)
plt.imshow(image_1_shifted,cmap='gray')
plt.title('image')
plt.subplot(1,2,2)
hist,bins=np.histogram(image_1_shifted.ravel(),255,[0,256])
plt.plot(hist)
plt.title('histogram')
plt.grid(True)
plt.show()
```

```python
hist, bins = np.histogram(image_1_shifted.ravel(), 256, [0, 256])
cumulative_hist = np.cumsum(hist)
plt.plot(cumulative_hist)
plt.xlabel('Gray Level')
plt.ylabel('Cumulative Frequency')
plt.title('Cumulative Histogram')
plt.grid(True)
plt.show()
```

# Equalization

```python
equ = cv2.equalizeHist(image_1)
plt.imshow(equ,cmap='gray')
plt.axis('off')
```

```python
plt.subplot(1,2,1)
plt.imshow(equ,cmap='gray')
plt.title('image')
plt.axis('off')
plt.subplot(1,2,2)
hist,bins=np.histogram(equ.ravel(),253,[0,256])
plt.plot(hist)
plt.title('histogram')
plt.grid(True)
plt.show()
```

```python
hist, bins = np.histogram(equ.ravel(), 256, [0, 256])
cumulative_hist = np.cumsum(hist)
plt.plot(cumulative_hist)
plt.xlabel('Gray Level')
plt.ylabel('Cumulative Frequency')
plt.title('Cumulative Histogram')
plt.grid(True)
plt.show()
```
